This is known as a Notebook environment for running commands written in the Python programming language in a web browser. This specific implementation is in Google's Collabatory environment, known as Colab. When you submit a command it is sent off to Google's servers where it runs and then send a response back to your browser.

To run a section of code, hover over the square brackets in the top left of a grey section (with programming stuff in it). The square brackets should turn into a 'Play' arrow (like on your video remote control). Click the arrow. It should start a little whirring motion and run the command. When the whirring stops the command is complete. If the arrow turns red there is an error in the code. We'll cross that bridge when we come to it.

Alternatively, you can click within the grey box itself and then press 'ctrl' and 'Enter' (on a Windows machine at least) at the same time to run the code.

The first command we're going to run will connect the Colab environment to your Google drive. Click this link, it will open a new browser tab to allow you to log in to Google Drive. You may get prompts asking for you to authorise Colab to access your files, please click Allow. Next to the numbers is a sort of double rectangle which you can click to copy the string to your clipboard. Then return to this windo and paste it in the "Enter you authorization code" box below. Press return and you'll be all set up. You should now see the text "Mounted at /content/gdrive".
